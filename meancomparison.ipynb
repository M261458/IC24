```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python


df = pd.read_csv('mydata.csv')
df = df[['SR_NDB','SR Food description','Nutrient_id','SR_Component','SR Mean per 100g','FF Mean per 100g']] # Fileter data frame to what we need.
stdevs0 = df.groupby('Nutrient_id',as_index=False)[['FF Mean per 100g']].std() # New data frame with Stdevs of FF mean
stdevs0 = stdevs0.rename(columns={"FF Mean per 100g": "stdev"}) # Rename FF mean column of new df to stdev.


stdevs_df = df.merge(stdevs0,on='Nutrient_id') # Merge; new df has both means and stdev
stdevs_df = stdevs_df.sort_values('SR Food description') # Sort by food type
stdevs_df = stdevs_df.reset_index() # reset index



stdev_list = []
for ind in stdevs_df.index: # calculates stdev difference for every row of df.
    stdev_list.append(abs((stdevs_df['FF Mean per 100g'][ind]-stdevs_df['SR Mean per 100g'][ind])/stdevs_df['stdev'][ind]))
stdevs_df['stdev_difference'] = stdev_list # Adds stdev differences to df.
# print(stdevs_df) # 'stdevs_df' has Food description, both means, stdevs, and stdev differences.
stdevs_df = stdevs_df.drop(['index'],axis=1)
```

```python
SR_adj_mean = []
FF_adj_mean = []

for ind in stdevs_df.index: # calculates stdev difference for every row of df.
    FF_adj_mean.append(abs(stdevs_df['FF Mean per 100g'][ind]/stdevs_df['stdev'][ind]))
    SR_adj_mean.append(abs(stdevs_df['SR Mean per 100g'][ind]/stdevs_df['stdev'][ind]))

stdevs_df['SR_adj_mean'] = SR_adj_mean
stdevs_df['FF_adj_mean'] = FF_adj_mean


# Add a Line where slope is m=1.
x_axis = [1,2,3,4,30]
y_axis = [1,2,3,4,30]
fig2 = px.line(x=x_axis,y=y_axis, width=970, height=700)
fig2.update_traces(line=dict(color = 'rgba(500,0,0,10)'))

fig1 = px.scatter(stdevs_df, x='SR_adj_mean', y='FF_adj_mean', labels={'SR_adj_mean':'SR Adjusted Mean','FF_adj_mean':'FF Adjusted Mean'}, hover_data=['SR Food description','SR_Component'],
                    trendline='ols', width=970, height=700, range_y=[0,14], range_x=[0,73]).add_trace(fig2.data[0])
                
fig1.show()



layout = {
    "xaxis": {
        "showticklabels": True,
        "showgrid": True,
        "zeroline": True,
        "linecolor": "black",
        "linewidth": 2,
        "mirror": True,
        "autorange": False,
        "range": [-1, 73],
        "title": 'SR Adjusted Mean'
    },
    "yaxis": {
        "showticklabels": True,
        "showgrid": True,
        "zeroline": True,
        "linecolor": "black",
        "linewidth": 2,
        "mirror": True,
        "autorange": False,
        "range": [-1, 14],
        "title": 'FF Adjusted Mean'
    },
}

fig3 = go.Figure(data=(fig1.data + fig2.data), layout=fig1.layout)
fig3.show()
```
